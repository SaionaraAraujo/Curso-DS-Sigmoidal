#Gráfico de linhas

* Princípio da conexão de Gestalt
> "A Gestalt diz que elementos que estão visualmente conectados são percebidos
como mais relacionados do que elementos sem conexão. Isso significa que 2 figuras idênticas em um layout, quando desconectadas, são vistas diferentemente do que as suas figuras disformes ligadas por uma linha. De todos os princípios que sugerem que os objetos estão relacionados, a conexão uniforme tende a ser o mais forte."
* Não faz sentido para variáveis categóricas
* Frequentemente o eixo X é uma unidade de tempo.


```python
# importar dados para dataframe
data = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/dengue-dataset.csv"
#o parâmetro parse_dates serve para ler a coluna de datas, que está no formato de String
df = pd.read_csv(data, index_col="data", parse_dates=True)
```

```python
df.head()
```

```python
df.tail()
```

##Plotar Gráfico

Vamos plotar os casos confirmados

```python
#criar o objeto fig e o ax
fig, ax = plt.subplots()

#selecionar a variável que será plotado e o tipo de gráfico
df["casos-confirmados"].plot(kind = "line", ax=ax)

#setar os títulos do gráfico e dos eixos
ax.set_title("Casos Confirmados")
ax.set_xlabel("Ano")
ax.set_ylabel("Quantidade")

#mostrar o gráfico
plt.show()
```

##Selecionar Intervalos

```python
#Plotar gráfico de linhas da temperatura média

fig,ax = plt.subplots()

#Slice do período, com a coluna que se quer plotar e o tipo de gráfico
# O método loc, acessa o grupo de valores passados no array, usando rótulos.
df.loc["2012-01-01":"2014-12-01","temperatura-media"].plot(kind = 'line', ax=ax)

#Definição dos títulos
ax.set_title ("Temperatura média por período")
ax.set_xlabel("Período")
ax.set_ylabel("Quantidade de Chuva")
plt.show()
```

##Plotar Múltiplos Gráficos Separadamente

```python
#Temperatura média e Casos Confrimados
#Neste caso, precisamos definir o objeto fig tenha mais de uma linha, sendo cada uma, um objeto ax
fig, (ax1,ax2) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,8))

#Gráfico da Temperatura
df.loc["2012-01-01":"2014-12-01","temperatura-media"].plot(kind = "line",ax=ax1)
ax1.set_title("Temperatura Média")
ax1.set_xlabel("Período")
ax1.set_ylabel("Temperatura")

#Gráfico dos Casos confirmados
df.loc["2012-01-01":"2014-12-01","casos-confirmados"].plot(kind = "line",ax=ax2)
ax2.set_title("Casos Confirmados")
ax2.set_xlabel("Período")
ax2.set_ylabel("Casos Confirmados")

plt.tight_layout()
```

##Plotar Gráficos  Múltiplos Sobrepostos


```python
#Normalizar o gráfico
df_norm = (df  - df.mean())/df.std()
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10,6))

#Temperatura Média
df_norm.loc["2012-01-01":"2014-12-01","temperatura-media"].plot(kind = "line", ax = ax)

#Casos Confirmados
df_norm.loc["2012-01-01":"2014-12-01","casos-confirmados"].plot(kind = "line", ax = ax)

#Titúlos
ax.set_title ("Temperatura Média x Casos Confirmados")
ax.set_xlabel("Período")
```

#Histograma

É um gráfico que tem por objetivo mostra a distribuição de frequencias, tendo no eixo Y a quantidade de ocorrências do obejto alvo da análise.

##Valores Absolutos

```python
#Plotar histograma da temperatura média

fig, ax = plt.subplots()

#Definir a coluna a ser plotada, a origem do dado e a quantidade de barras
ax.hist('temperatura-media', data = df, bins = 10)

#títulos
ax.set_title("Histograma Temperatura Média")
ax.set_xlabel("Temperatura")
ax.set_ylabel("Ocorrências")

plt.tight_layout()
```

##Valores Relativos

O histograma acima está representando as ocorrências os valores absolutos, para a visão dos valores relativos, é preciso inserir o parâmetro `density=True`:

```python
#Plotar histograma com valores relativos da temperatura média

fig, ax = plt.subplots()

#Definir a coluna a ser plotada, a origem do dado e a quantidade de barras
ax.hist('temperatura-media', data = df, bins = 40, density=True)

#títulos
ax.set_title("Histograma Temperatura Média")
ax.set_xlabel("Temperatura")
ax.set_ylabel("Ocorrências")

plt.tight_layout()
```

##Plotar o gráfico de pizza

```python
df3 = pd.read_csv ('https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv', index_col = "codigo_ocorrencia")

df3.head()
```

Para gerar o gráfico de pizza, vamos transformar a coluna de classificação em um dicionário, contando a ocorrência de cada uma:

```python
ocorrencia_classificaco = dict(df3.ocorrencia_classificacao.value_counts())

#Separar os labels e valores
label = ocorrencia_classificaco.keys()
value = ocorrencia_classificaco.values()

explode = (0.1, 0.1, 0.1)
```

```python
#plotar
fig,ax = plt.subplots(figsize= (6,6))

ax.pie(value, labels=label)
ax.set_title("Classificação das Ocorrências")

plt.tight_layout()
```

```python
#plotar com o explode e sombra
fig,ax = plt.subplots(figsize= (6,6))

#o parâmetro explode afasta as fatias e o shadow plota uma sombra
ax.pie(value, labels=label,explode=explode, shadow=True)
ax.set_title("Classificação das Ocorrências")

plt.tight_layout()
```

#Nuvem de Parlavras

Indicado para análise de texto.

```python
#Instalar a bibliotec

!pip install wordcloud -q
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

df6 = pd.read_csv("http://data.insideairbnb.com/united-kingdom/england/greater-manchester/2021-12-23/visualisations/listings.csv")
```

```python
df6.head()
```

```python
#Eliminar valores ausenste da coluna nome

#contar
df6.name.isnull().sum()
```

##Juntar as palavras

É preciso juntaras palavras que que seja possível gerar a nuvem com todas as entradas da coluna:

```python
#concatenar as palavras
palavras = " ".join(s for s in df6.name)

len(palavras)

print(palavras)
```

No comando acima, vimos que há muitas palavras que podemos excluir para não "poluir" nossa nuvem de palavras. Vamos fazê-lo a seguir:

##Primeira versão: sem a limpeza das palavras

```python
#Vamos atualizar esta lista com as palabvras que queremos excluir
stopwords = set(STOPWORDS)

#Gerar uma nuvem de palvras:
nuvem = WordCloud(stopwords=stopwords,
                  background_color="gray",).generate(palavras)

#Mostrar imagem:
fig, ax = plt.subplots(figsize=(10,6))

ax.imshow(nuvem, interpolation="bilinear")

plt.show()
```

##Limpeza das palavras para gerar a nuvem

```python
#palavras que queremos tirar
stopwords.update(['1', '2', '3', '4', '5', '6', '7', '8', '9', '0', 'with', 'bar', 'and', 'en'])
#Gerar uma nuvem de palvras:
nuvem1 = WordCloud(stopwords=stopwords,
                  background_color="black",).generate(palavras)

#Mostrar imagem:
fig, ax = plt.subplots(figsize=(16,8))

ax.imshow(nuvem1, interpolation="bilinear",)
ax.set_axis_off()

plt.imshow(nuvem1)
```
